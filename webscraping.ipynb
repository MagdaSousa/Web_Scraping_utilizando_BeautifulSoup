```
import pandas as pd
```

```
from urllib.request import urlopen
from bs4 import BeautifulSoup
url ='https://stats.nba.com/team/1610612741/'
# aqui tento acessar a url
try:
    html = urlopen(url)
except:
    print("An error occured.")
```

```
# Criar o objeto BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
print(soup.title)# verificando o título da página do site da NBA
```

#Pegando Texto de uma Página Web
Neste ponto vou clicar com o botão direito na página web a qual quero obter os dados, e clicar em inspecionar elemento, e vou procurar as tags, relacionadas ao que eu quero extrair do site.Como nesse caso quero extrair os times da MBA, estão estou procurando as tag que contem o link e os nomes que quero resgatar.image.png

```
# Pegar todo o texto da div columns / small-12 / section-view-overlay
jogadores_chicabo_bulls = soup.find(class_='columns / small-12 / section-view-overlay')
```

Agora pretendo acessar o nome dos jogadores da equipe do chigado bullsimage.png

```
# Pegar o texto de todas as instâncias da tag <a> dentro da div columns / small-12 / section-view-overlay
name_list_items = jogadores_chicabo_bulls.find_all('a')
print(name_list_items)
```

```
jogadores = jogadores_chicabo_bulls.find_all('data-ng-repeat')
print(jogadores)
```
