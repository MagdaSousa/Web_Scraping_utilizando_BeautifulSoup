```
import bs4 as bs
import urllib.request
url ='https://stats.nba.com/team/1610612741/'
# aqui tento acessar a url
sauce = urllib.request.urlopen(url).read()
soup= bs.BeautifulSoup(sauce,'lxml')
print(soup.title)
print(soup.title.name)#nome da tag
print(soup.title.string)#nome da páagina sem a tag
print(soup.title.text)#tbm mostra o nome sem a tag
print(soup.p)# Me parece ser o parágrafo
print(soup.find_all('p'))# todos os parágrafos
```

```
for paragrafo in soup.find_all('p'):
  print(paragrafo)# aqui consigo mostrar todos os paragrafos linha a linha junto com as tags
```

```
for paragrafo in soup.find_all('p'):
  print(paragrafo.text)#mostrando os paragrafos um a um sem as tags
```

```
for url_2 in soup.find_all('a'):
  print(url_2.get('href'))# extraindo os links da pagina
```

```
 nav = soup.nav
 print(nav)# o primeiro nav que ele encontrar
```

```
for nav_text in nav.find_all('a'):
  print(nav_text.get('href'))# extraindo os links do nav bar
```

```
body = soup.body
for paragrafo in body.find_all('p'):# pega todos os parágrafos do corpo da página, mas aqui aparentemente ele só considera o rodapé com parágrafo
  print(paragrafo.text)
```

```
for div in soup.find_all('div'):
  print(div.text)
```

```
for div2 in soup.find_all('div',class_='body'):
    print(div2.text)
```
