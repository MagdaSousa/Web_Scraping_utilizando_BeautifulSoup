```
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```
url='https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol'
response =requests.get(url)
```

```
soup= BeautifulSoup(response.text,'html.parser')
tables =soup.find_all('table')
len(tables)
```

```
#neste caso estou acessando todas as tabelas existentes na página, cada tabela possui um id diferente começando em 0, esse id eu passo
#para fazer o slice novas_tabelas[0],etc
novas_tabelas=pd.read_html(str(tables),header=0)
novas_tabelas[3]
```

```
Participacoes_dos_clubes =pd.DataFrame(novas_tabelas[10])
Participacoes_dos_clubes
```

```

import plotly.express as px


fig = px.bar(Participacoes_dos_clubes, x="Clube", y="Participações na Taça Brasil(1959–1968)", title="Participações dos clubes na taça Brasil")
fig.show()
```

```
fig = px.bar(Participacoes_dos_clubes, x="Clube", y="Total de participações(1959–2020)", title="Total de participações(1959–2020)")
fig.show()
```
