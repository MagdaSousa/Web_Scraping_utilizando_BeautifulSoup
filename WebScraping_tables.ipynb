```
import bs4 as bs
import urllib.request
import pandas as pd
url ='https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/2020'
# aqui tento acessar a url
sauce = urllib.request.urlopen(url).read()
soup= bs.BeautifulSoup(sauce,'lxml')
print(soup.title)
```

```
table = soup.table
print(table)
```

```
# table = soup.find('table')
# print(table)
```

```
table_rows= table.find_all('tr')

for tr in table_rows:
  td= tr.find_all('td')
  row =[i.text for i in td]
  print(row)
```

```
dfs = pd.read_html(url,header=0)
for df in dfs:
  print(df)
  #fazendo o mesmo acesso ao html só que com pandas
```

Acessando o xml da página

```
url_xml ='https://www.cbf.com.br/sitemap.xml'
# aqui tento acessar a url
sauce = urllib.request.urlopen(url_xml).read()
soup= bs.BeautifulSoup(sauce,'xml')
print(soup)
```

```
for url in soup.find_all('loc'):
  print(url.text)
```
